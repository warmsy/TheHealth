```python
from google_play_scraper import Sort, reviews

# 처음부터 200개 리뷰 가져오기
first_200_reviews, _ = reviews(
    'com.samsunglife.android.app.thehealth',
    lang='ko',
    country='kr',
    count=200,  # 처음 200개 리뷰만 가져오기
    sort=Sort.NEWEST  # 최신순으로 정렬
)

all_reviews = first_200_reviews

# 다음 페이지부터 시작하여 나머지 리뷰 가져오기
page_count = 2
while True:
    next_reviews, _ = reviews(
        'com.samsunglife.android.app.thehealth',
        lang='ko',
        country='kr',
        count=200,
        sort=Sort.NEWEST,
        pagination_token=first_200_reviews[-1]['reviewId']  # 이전 페이지의 마지막 리뷰의 reviewId 사용
    )
    if not next_reviews:
        break  # 더 이상 리뷰가 없으면 종료
    all_reviews.extend(next_reviews)
    page_count += 1
    print(f"Fetched page {page_count}")

print(f"Total reviews fetched: {len(all_reviews)}")
```
